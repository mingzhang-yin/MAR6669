```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the mixture model
mu1 = 100
sigma1 = 10
mu2 = 30
sigma2 = 15
weight1 = 0.3
weight2 = 1 - weight1

# Generate data points from the mixture model
num_samples = 1000
z = np.random.binomial(1, weight1, num_samples)
x = z * np.random.normal(mu1, sigma1, num_samples) + (1 - z) * np.random.normal(mu2, sigma2, num_samples)


plt.figure(figsize=(10, 6))
# Plot the histogram of the data points
plt.hist(x, bins=30, density=True, alpha=0.5)

# Plot the true density of the mixture model
x_range = np.linspace(-20, 120, 1000)
true_density = weight1 * norm.pdf(x_range, mu1, sigma1) + weight2 * norm.pdf(x_range, mu2, sigma2)
plt.plot(x_range, true_density, 'k--', label='Distribution of handbag price (\$) in the store')
ax = plt.gca()
ax.set_yticks([])

# Show the plot
plt.legend(fontsize=16, loc=1)
plt.show()
```
